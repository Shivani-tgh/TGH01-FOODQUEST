```python
%sql
CREATE TABLE fq_dev_catalog.silver.dim_country
(
    sys_id STRING NOT NULL,
    country_code STRING NOT NULL,
    country_name STRING NOT NULL,
    currency_code STRING,
    region STRING,
    currency STRING,
    timezone STRING NOT NULL,
    is_active BOOLEAN NOT NULL,
    source_system STRING NOT NULL,

    CONSTRAINT country_pkey PRIMARY KEY (country_code)
    
)
USING DELTA
TBLPROPERTIES (
    delta.enableChangeDataFeed = true
);
```

```python
country_df=spark.read.table("fq_dev_catalog.silver.dim_country")
```

```python
df_new_country=spark.createDataFrame([
    ("AE", "United Arab Emirates", "AED", "Middle East", "UAE Dirham", "Asia/Dubai", True, "FQ")], 
    ["country_code", "country_name", "currency_code", "region", "currency", "timezone", "is_active", "source_system"])
```

```python
from pyspark.sql.functions import expr
df_new_country = df_new_country.withColumn("sys_id", expr("uuid()"))
```

```python
df_country_final=df_new_country.select("sys_id","country_code","country_name","currency_code","region","currency","timezone","is_active","source_system")
```

```python
df_country_final.display()
```

```python
df_country_final.printSchema()
```

```python
%sql
describe table fq_dev_catalog.silver.dim_country
```

```python
df_country_final.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("fq_dev_catalog.silver.dim_country")
```
